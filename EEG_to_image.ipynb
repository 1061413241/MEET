## Generate brain images based on SEED dataset

```python
import scipy.io as sio
from Utils_Bashivan import *
import numpy as np
import os


experiment1 = ['1_20131027.mat', '2_20140404.mat', '3_20140603.mat', '4_20140621.mat', '5_20140411.mat', 
                '6_20130712.mat', '7_20131027.mat', '8_20140511.mat', '9_20140620.mat', '10_20131130.mat', 
                '11_20140618.mat', '12_20131127.mat', '13_20140527.mat', '14_20140601.mat', '15_20130709.mat']

experiment2 = ['1_20131030.mat', '2_20140413.mat', '3_20140611.mat', '4_20140702.mat', '5_20140418.mat', 
                '6_20131016.mat', '7_20131030.mat', '8_20140514.mat', '9_20140627.mat', '10_20131204.mat', 
                '11_20140625.mat', '12_20131201.mat', '13_20140603.mat', '14_20140615.mat', '15_20131016.mat']

experiment3 = ['1_20131107.mat', '2_20140419.mat', '3_20140629.mat', '4_20140705.mat', '5_20140506.mat', 
                '6_20131113.mat', '7_20131106.mat', '8_20140521.mat', '9_20140704.mat', '10_20131211.mat', 
                '11_20140630.mat', '12_20131207.mat', '13_20140610.mat', '14_20140627.mat', '15_20131105.mat']

time_len = 2

for cnt in range(0, 15):
    ## Load signal
    # Experiment-1
    flag = 1
    data = sio.loadmat('./dataset/SEED/ExtractedFeatures/' + experiment1[cnt])   # Change to your dataset path
    label = []
    session1_label = [1,0,-1,-1,0,1,-1,0,1,1,0,-1,0,1,-1]
    for i in range(1,16):
        path = 'de_movingAve' + str(i)
        feat_i = data[path]
        num = int(feat_i.shape[1]/time_len)
        for k in range(num):
            label.append(session1_label[i-1]+1)
            feat_time = feat_i[:,k*time_len:(k+1)*time_len,:]
            feat_time = np.expand_dims(feat_time, axis=0)
            if flag == 1:
                feats = feat_time
                flag = 0
            else:
                feats = np.concatenate((feats, feat_time), axis=0)



    # Experiment-2
    data = sio.loadmat('./dataset/SEED/ExtractedFeatures/' + experiment2[cnt])
    session2_label = [1,0,-1,-1,0,1,-1,0,1,1,0,-1,0,1,-1]
    for i in range(1,16):
        path = 'de_movingAve' + str(i)
        feat_i = data[path]
        num = int(feat_i.shape[1]/time_len)
        for k in range(num):
            label.append(session2_label[i-1]+1)
            feat_time = feat_i[:,k*time_len:(k+1)*time_len,:]
            feat_time = np.expand_dims(feat_time, axis=0)
            feats = np.concatenate((feats, feat_time), axis=0)

    # Experiment-3
    data = sio.loadmat('./dataset/SEED/ExtractedFeatures/' + experiment3[cnt])
    session3_label = [1,0,-1,-1,0,1,-1,0,1,1,0,-1,0,1,-1]
    for i in range(1,16):
        path = 'de_movingAve' + str(i)
        feat_i = data[path]
        num = int(feat_i.shape[1]/time_len)
        for k in range(num):
            label.append(session3_label[i-1]+1)
            feat_time = feat_i[:,k*time_len:(k+1)*time_len,:]
            feat_time = np.expand_dims(feat_time, axis=0)
            feats = np.concatenate((feats, feat_time), axis=0)

    # Reshape
    label = np.expand_dims(label, axis=1)
    feats = np.reshape(feats, (-1,310))

    locs = sio.loadmat('./dataset/SEED/3DLocs.mat')  # the path of 3D-location file
    locs_3d = locs['locs']
    locs_2d = []

    # 3D locations convert to 2D
    for e in locs_3d:
        locs_2d.append(azim_proj(e))

    images_timewin = np.array(gen_images(np.array(locs_2d), feats, 32, normalize=True))
    images_timewin = np.reshape(images_timewin, (-1,time_len,5,32,32))
    sio.savemat("./dataset/SEED/eeg_convert_to_image/S" + str(cnt+1) + "_32@32.mat", {"img":images_timewin, 'label':label})

```
